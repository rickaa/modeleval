```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets
```

```python
diabetes = datasets.load_diabetes()
```

```python
# Sperate train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(diabetes.data, diabetes.target, test_size=0.2, random_state=0)
```

```python
# 1. Set up the model
model = LinearRegression()
# 2. Use fit
model.fit(X_train, y_train)
```

```python
from modeleval import RegressionEvaluator
```

```python
# Set up a regression evaluator object
eva = RegressionEvaluator()
```

```python
eva.evaluate(model, X_test, y_test, plot=True, save=True, save_folder= "result/")
```

```python
# build another model
from sklearn.linear_model import Lasso
```

```python
model2 = Lasso()
```

```python
model2.fit(X_train, y_train)
```

```python
eva.evaluate(model2, X_test, y_test)
```

```python
# find the model with maximum r square
bestmodel_r2 = eva.find_best_model([model, model2], X_test, y_test, "r2")
```

```python
# find the model with minimum rmse
bestmodel_rmse = eva.find_best_model([model, model2], X_test, y_test, "rmse")
```
